```
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```
img_path = "./test.jpg"
# OpenCV의 삼원색 기본 순서는 BGR이다.
img_bgr = cv.imread(img_path,1)
img_rgb = cv.cvtColor(img_bgr ,cv.COLOR_BGR2RGB)
cv2_imshow(img_bgr)
cv2_imshow(img_rgb)
```

```
b,g,r = cv.split(img_bgr)
print(b,g,r)
img_rbg = cv.merge((r,b,g))
cv2_imshow(img_rbg)
```
