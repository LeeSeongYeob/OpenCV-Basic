#Image 출력 및 저장
- Colab 환경에서 실행하기 위해서
    - from google.colab.patches import cv2_imshow

```
import cv2 as cv
from google.colab.patches import cv2_imshow
```

- flag 값 
    - 1 = color
    - 0 = grayscale
    - -1= alpha 채널

```
img_path = "./test.jpg"
img3 = cv.imread(img_path,flags = 1)
cv2_imshow(img3)
```

```
img3 = cv.imread(img_path,flags = 0)  # gray scale
cv2_imshow(img3)
```

```
cv.waitKey(0)   # key 입력이 있을 때 까지 대기
cv.imwrite('./test.jpg',img3) 
cv.imwrite('./test.png',img3)
cv.imwrite('./test.bmp',img3)
```
